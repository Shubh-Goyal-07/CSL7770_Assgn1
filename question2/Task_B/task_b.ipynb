```python
import librosa
from librosa.filters import get_window
from scipy.signal import stft
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
WINDOW_SIZE = 1024
OVERLAP = 512
```

```python
def plot_spectrogram(file_path):
    waveform, sr = librosa.load(file_path, sr=None)
    window = get_window('hamming', WINDOW_SIZE, fftbins=True)
    _, _, Zxx = stft(waveform, window=window, nperseg=WINDOW_SIZE, noverlap=OVERLAP, fs=sr)
    spectrogram = np.abs(Zxx)

    plt.figure(figsize=(16, 6))
    librosa.display.specshow(librosa.amplitude_to_db(spectrogram, ref=np.max), y_axis='log', x_axis='time', sr=sr)
    plt.colorbar(format='%+2.0f dB')
    plt.title('Spectrogram')
    plt.tight_layout()
    plt.show()
```

```python
plot_spectrogram('songs_data/perfect_ed_pop.mp3')
```

```python
plot_spectrogram('songs_data/love_the_way_u_lie_eminem_rap.mp3')
```
