```python
import matplotlib.pyplot as plt
```

The Train and Test Accuracies were recorded during the training of the model. They can be found in log files.

```python
models = ['cnn', 'nn', 'dt', 'rf', 'svm', 'knn']

hamming_train = [97.9, 92.25, 99.15, 100.0, 79.5, 76.9]
hamming_test = [92.0, 92.25, 54.77, 70.80, 62.44, 63.4]

hann_train = [97.9, 79.64, 98.4, 100.0, 78.95, 71.66]
hann_test = [90.5, 79.65, 48.25, 61.82, 61.3, 56.95]

rect_train = [97.8, 73.00, 98.6, 100.0, 77.62, 70.72]
rect_test = [93.0, 73.01, 49.6, 62.49, 61.53, 57.18]
```

```python
plt.figure(figsize=(10, 5))
plt.plot(models, hann_train, label='Train', marker='o')
plt.plot(models, hann_test, label='Test', marker='o')
plt.title('Hann Window')
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(models, hamming_train, label='Train', marker='o')
plt.plot(models, hamming_test, label='Test', marker='o')
plt.title('Hamming Window')
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(models, rect_train, label='Train', marker='o')
plt.plot(models, rect_test, label='Test', marker='o')
plt.title('Rectangular Window')
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(models, hamming_train, label='Hamming Train', color='blue', marker='o', alpha=0.5)
plt.plot(models, hamming_test, label='Hamming Test', linestyle='--', color='blue', marker='o', alpha=0.5)
plt.plot(models, hann_train, label='Hann Train', color='orange', marker='o', alpha=0.5)
plt.plot(models, hann_test, label='Hann Test', linestyle='--', color='orange', marker='o', alpha=0.5)
plt.plot(models, rect_train, label='Rectangular Train', color='green', marker='o', alpha=0.5)
plt.plot(models, rect_test, label='Rectangular Test', linestyle='--', color='green', marker='o', alpha=0.5)
plt.title('Train and Test Accuracy')
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
